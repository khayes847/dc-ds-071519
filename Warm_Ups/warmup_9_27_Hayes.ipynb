```python
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn.cluster import AgglomerativeClustering
from sklearn.neighbors import KernelDensity
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline  
```

```python
X, y = make_blobs(n_samples=1000, n_features=5, centers=7, cluster_std=3.5)

plt.scatter(X[:, 0], X[:, 1], c = y, s = 25)
plt.title('Clusters without centers')
plt.show()

k_means = KMeans(n_clusters = 7)
k_means.fit(X)
y_hat = k_means.predict(X)

plt.scatter(X[:, 0], X[:, 1], c = y_hat, s = 25)
cl_centers = k_means.cluster_centers_
plt.scatter(cl_centers[:, 0], cl_centers[:, 1], c='black', s=100);
plt.title('Clusters with centers')
plt.show
```

3. Advantages: In general, they're easy to make and to adapt to new datasets, they can be used with large datasets, and they can be used with groups of different shapes (not just circles). In addition, they can be used when you don't necessarily have a defined target variable. However, they do not adjust to different-sized clusters well, they are influenced heavily by outliers, and the distance between clusters in datasets with too many dimensions can converge, necessitating PCA.

4. Agglomerative hierarchical clustering is when seperate objects with their own clusters are grouped together to form larger clusters, until they are merged into a parent cluster. Divisive hierarchical clustering is the opposite, in which parent clusters are broken down into their seperate elements.

5.

```python
X, y = make_blobs(n_samples= 100, n_features=5, centers=7, cluster_std=1.5)
plt.scatter(X[:, 0], X[:, 1], c = y, s = 10);
plt.title('Unagglomerated hierarchical clusters')
plt.show()
```

```python
agg_comp = AgglomerativeClustering(linkage ="complete", n_clusters=6)
agg_avg = AgglomerativeClustering(linkage ="average", n_clusters=6)
agg_ward = AgglomerativeClustering(linkage ="ward", n_clusters=6)

as_comp = agg_comp.fit_predict(X)
as_avg = agg_avg.fit_predict(X)
as_ward = agg_ward.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = as_comp, s = 10);
plt.title('Linkage complete')
plt.show()

plt.scatter(X[:, 0], X[:, 1], c = as_avg, s = 10);
plt.title('Linkage average')
plt.show()

plt.scatter(X[:, 0], X[:, 1], c = as_ward, s = 10);
plt.title('Linkage ward')
plt.show()
```

```python
k_means = KMeans(n_clusters = 7)
k_means.fit(X)
y_hat = k_means.predict(X)

plt.scatter(X[:, 0], X[:, 1], c = y_hat, s = 10)
cl_centers = k_means.cluster_centers_
plt.scatter(cl_centers[:, 0], cl_centers[:, 1], c='black', s=40);
plt.title('Kmeans clustering')
plt.show()
```
